```python
import pandas as pd
import numpy as np
```

```python
webster_cencus = pd.read_csv("/Users/brandonyu/Downloads/Blockgroups to ESAA Spreadsheet with student counts - Webster ESAA Census (1).csv")
webster_data = pd.read_csv("/Users/brandonyu/Downloads/Blockgroups to ESAA Spreadsheet with student counts - Webster BlockGroup Data.csv")
```

```python
webster_cencus.head()
```

```python
webster_cencus = webster_cencus.drop(columns=["E_AA_NAME", "Count of current SFUSD Kindergarten students", "Unnamed: 23"])
```

```python
webster_cencus.head()
```

```python
webster_cencus = webster_cencus.iloc[1:, :]
```

```python
webster_cencus.head()
```

```python
webster_cencus["Block Group"] = webster_cencus["Label (Grouping)"].str.extract(r'Group\s(\d)').astype(int)
```

```python
webster_cencus["Block Group"]
```

```python
webster_cencus["Cencus Tract"] = webster_cencus["Label (Grouping)"].str.extract(r'Tract\s(\d+\.*\d+)').astype(float)
```

```python
webster_cencus["Cencus Tract"]
```

```python
columns = webster_cencus.columns.tolist()
reordered_columns = [columns[-2], columns[-1]] + columns[:-2]
webster_cencus = webster_cencus[reordered_columns]
webster_cencus = webster_cencus.drop(columns=["Label (Grouping)"])
webster_cencus.head()
```

```python
webster_data.head()
```

```python
webster_data = webster_data.iloc[1:, 2:]
```

```python
webster_data.head()
```

```python
webster_total = webster_cencus.merge(webster_data, left_on="Geoid1", right_on="Geoid1")
```

```python
webster_total.head()
```

```python
cencus_tracts = pd.read_csv("/Users/brandonyu/Downloads/region_2020_censustract_9128033371162871599.csv")
```

```python
cencus_tracts.head()
```

```python
print(webster_total["Geoid1"].dtype)
print(cencus_tracts["geoid"].dtype)
```

```python
webster_total["Geoid1"] = webster_total["Geoid1"].astype(str).str[:10]
```

```python
cencus_tracts["geoid"]
```

```python
print(webster_total["Geoid1"].dtype)
print(cencus_tracts["geoid"].dtype)
```

```python
webster_total["Geoid1"] = webster_total["Geoid1"].astype(int)
```

```python
geo_total = webster_total.merge(cencus_tracts, left_on="Geoid1", right_on="geoid")
```

```python
geo_total.head()
```

```python
geo_total["Weighted HH Inc Int"] = geo_total["Weighted HH Inc"].str[1:].str.split(',')
geo_total["Weighted HH Inc Int"] = geo_total["Weighted HH Inc Int"].apply(lambda lst: int(''.join(map(str, lst))))
geo_total["Weighted HH Inc Int"]
```

```python
geo_total.head()
```

```python
print(geo_total["Weighted HH Inc"].dtype)
```

```python
weighted_median_bool = geo_total[geo_total['Weighted HH Inc Int'] != 0]
weighted_median = int(weighted_median_bool['Weighted HH Inc Int'].median())
weighted_median
```

```python
geo_total['Weighted HH Inc Int'] = geo_total['Weighted HH Inc Int'].replace(to_replace=0, value=weighted_median)
```

```python
geo_total['Weighted HH Inc Int']
```

```python
geo_total["Median HH Income_x"] = geo_total["Median HH Income_x"].fillna("$0")
```

```python
geo_total.head()
```

```python
geo_total["Median HH Inc Int"] = geo_total["Median HH Income_x"].str[1:].str.split(',')
geo_total["Median HH Inc Int"] = geo_total["Median HH Inc Int"].apply(lambda lst: int(''.join(map(str, lst))))
```

```python
geo_total.head()
```

```python
geo_total["Median HH Inc Int"]
```

```python
median_median_bool = geo_total[geo_total['Median HH Inc Int'] != 0]
median_median = int(median_median_bool['Median HH Inc Int'].median())
median_median
```

```python
geo_total['Median HH Inc Int'] = geo_total['Median HH Inc Int'].replace(to_replace=0, value=median_median)
```

```python
geo_total.head()
```

```python
columns = geo_total.columns.tolist()
last_two_columns = columns[-2:]
remaining_columns = columns[:-2]
new_order = remaining_columns[:2] + last_two_columns + remaining_columns[2:]
geo_total = geo_total[new_order]
geo_total.head()
```

```python
geo_total.drop(columns=["Weighted HH Inc", "Median HH Income_x", "Total: (Population Race P8 Census 2020.csv)", "name_", "geoid"])
```
